```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#laoding the dataset to a panda dataframe
data=pd.read_csv('/content/AIDS_Classification.csv')
```

```python
#splitting features and target
X=data.drop(columns='infected',axis=1)
Y=data['infected']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Print shapes of the datasets
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Initializing the Random Forest model
model = RandomForestClassifier(random_state=42)
```

```python
# Training the Random Forest model with training data
model.fit(X_train, Y_train)
```

```python
#ACCURACY ON TRAINING DATA
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data:', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data:', test_data_accuracy)
```

```python
input_data = (1002,3,61,49.4424,0,0,0,90,0,1,895,0,0,1,3,0,1,0,162,218,392,564)

# Convert the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array for a single prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Make a prediction
prediction = model.predict(input_data_reshaped)
print("Prediction:", prediction)

# Check the result
if prediction[0] == 0:
    print('The person does not have AIDS')
else:
    print('The person has AIDS')
```
